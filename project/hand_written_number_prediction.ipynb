```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
(x_train,Y_train),(x_test,Y_test) = keras.datasets.mnist.load_data()
```

```python
len(x_train)
```

```python
modal = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation='softmax')
])
```

```python
x_train[:5].shape
```

```python
x_train[4]
```

```python
Y_train[:5]
```

```python
x_test[0]
```

```python
Y_test[0]
```

```python
modal.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
x_train_flat=x_train.reshape(-1,28*28)
x_train_flat=x_train_flat/255
x_test_flat=x_test.reshape(-1,28*28)
```

```python
modal.fit(x_train_flat,Y_train,epochs=5)
```

```python
A=20
```

```python
plt.matshow(x_train[A])
```

```python
predicted=modal.predict(x_test_flat)
predicted[A]
```

```python
np.argmax(predicted[A])
```

```python
cm=tf.math.confusion_matrix(labels=Y_test,predictions=predicted.argmax(axis=1))
cm
```

```python
import seaborn as sns
plt.figure(figsize=(10,10))
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

When reshaping data, particularly for neural networks, you often need to transform a 2D image (e.g., 28x28 pixels) into a 1D vector (e.g., 784 pixels). The `reshape` method in NumPy allows you to do this.

For example, if you have an array `x` with shape `(num_images, height, width)`, you can reshape it to `(num_images, height * width)` using `x.reshape(-1, height * width)`.

*   The `-1` tells NumPy to automatically calculate the size of that dimension based on the other dimensions and the total number of elements in the array. In this case, it will infer the number of images.
*   `height * width` calculates the total number of pixels in each image, effectively flattening it into a single vector.

This is why the code `x_train.reshape(-1, 28*28)` and `x_test.reshape(-1, 28*28)` works correctly for both training and test datasets, regardless of their differing number of samples.
